```python
import streamlit as st
import pandas as pd

# 제목 설정
st.title('지역별 의사 수 시각화')

# Excel 파일 로드
df = pd.read_excel('test_2.xlsx')

# 데이터 확인을 위한 테이블 표시
st.write("지역별 의사 수 데이터", df)

# 간단한 바 차트로 시각화
st.bar_chart(df.set_index('지역')['의사'])
```
