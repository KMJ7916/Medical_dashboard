```python
import matplotlib.pyplot as plt

# 데이터 준비
labels = ['사과', '바나나', '딸기', '오렌지', '포도']
sizes = [215, 130, 245, 210, 120]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']

# 도넛 차트 그리기
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # 원형을 유지하기 위한 설정

# 중앙에 원 그리기로 도넛 차트 형태 완성
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('과일 선호도 조사 결과')  # 차트 제목 추가
plt.show()
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# 한글폰트 경로 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# Streamlit 앱의 제목 설정
st.title("의료종사자 수")

# 데이터 가져오기
df = pd.read_csv("ttest.csv")

# 데이터 정제

# 데이터 분석
# 이 부분에서는 각 자치구별로 의사 수를 집계합니다.
df_analysis = df_selected.groupby("자치구별").sum().reset_index()


# 만약 자세한 조정이 필요하다면, Matplotlib를 사용하여 그래프를 그리고
# Streamlit에서 그래프를 표시할 수 있습니다. 예를 들면 아래와 같습니다.

fig, ax = plt.subplots()
ax.bar(df_analysis["자치구별"], df_analysis["의사"])
ax.set_xlabel("자치구별")
ax.set_ylabel("의사 수")
ax.set_title("자치구별 의료종사자(의사) 수")
st.pyplot(fig)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'

# 샘플 데이터 생성
data = {
    '이름': ['김철수', '이영희', '박민수', '최영미', '장민철'],
    '점수': [85, 92, 88, 95, 90]
}

# pandas DataFrame 생성
df = pd.DataFrame(data)

# 막대 그래프 그리기
df.plot(kind='bar', x='이름', y='점수', color='skyblue')

plt.title('학생별 점수')  # 그래프 제목
plt.xlabel('이름')         # x축 라벨
plt.ylabel('점수')         # y축 라벨
plt.xticks(rotation=0)    # x축 라벨 회전 각도
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# data.xlsx 파일 불러오기
df = pd.read_excel('data_file/medical_person.xlsx', index_col='지역')

df[['의사', '치과의사']].plot(kind='bar', figsize=(15, 7))

plt.title('지역별 의료 인력 비교')
plt.xlabel('지역')
plt.ylabel('인력 수')
plt.xticks(rotation=45)
plt.legend(title='직종')

plt.tight_layout()
plt.show()
```
